## Learning Gradient Descent

$\Delta C \approx \nabla C \cdot \Delta v$
This equation associates the change in Cost, gradient, and the change in parameter vector v.

We want to make $\Delta C$ negative, we can achieve this by choosing 
$$ \Delta v = -\eta \nabla C
$$
where $\eta$ is a small, positive paramter known as the learning rate.
This will yield the equation $$
\Delta C \approx -\eta \nabla C \cdot \nabla C = -\eta ||\nabla C||^2
$$ 
and since $||\nabla C||^2 \geq 0$, therefore, this guarantees that $\Delta C \leq 0$

for a ball example, we will compute a value for the change in the ball's position moving down the value, dnoted as $\Delta v$, then the ball's position $v$ will move by that amount:
$$
v \rightarrow v' = v - \eta \nabla C
$$

To sum up, gradient dfescent algorithm works is to repeatedly compute the gradient $\nabla C$, and then move in the opposite direction, falling down the valley

To make gradient descent work correctly, we need to choose the leanring rate $\eta$ to be small enough that equation is a good appromixation. If we don't, we might ned up with $\Delta C > 0$, which means the cost function is not decreasing, but increasing, which is not helpful when it comes to minizing its values.

Gradient descent function works the same even when C is a function of many more variables. Supose that $C$ is a function of m vairables, $v_1, v_2, ..., v_m$. Then the change $\Delta C$ in $C$ produced by a small change $\Delta v = (\Delta v_1, ... \Delta v_m)^T$ is 
$$ \Delta C = \nabla C \cdot \Delta v
$$
where the garidient $\nabla C$ is the vector
$$
\Delta C \equiv (\frac{\delta C}{\delta v_1}, ..., \frac{\delta C}{\delta v_m})^T
$$

Just as for the two variables case, we can choose
$$
\nabla \equiv -\eta \nabla C
$$
we are guaranteed that $\Delta C$ will be negative. This agains give us a way of following the graidnent to a minimum, even when $C$ is a function of many variables, by repeatedly applying the update rule
$$
v \rightarrow v' = v - \eta \nabla C
$$
(Think of this in 2D space, when we want to find the min of y, we would decrease the value of X step by step in the negative gradient direction, until the value is y is locally minimized).

This update rule gives us a way of repeatedly chanigng the position of all the vetors $v$ in rder to find a min of our cost function $C$. This rule doesn't always work, but often works extremely well in practice.

So gradient descent can be viewed as a way of taking small steps in the direction which does the most to immediately decrease $C$.

### How is gradient descent related to Neural Network

our "position" now has components $w_k$ and $b_l$
, and the gradient vector C  has corresponding components C/wk
 and C/bl
. Writing out the gradient descent update rule in terms of components, we have

$$
w_k \rightarrow w'_k = w_k - \eta \frac{\delta C}{\delta w_k}
$$
$$
b_l \rightarrow b'_l = b_l - \eta \frac{\delta C}{\delta b_l}
$$
By repeatedly applying this update rule we can "roll down the hill", and hopefully find a minimum of the cost function. 

There are a number of challenges in the graident descent rule. We will go over one problem for now: the cost function has the form $C_x \equiv \frac{||y(x) - \alpha||^2}{2}$ for individual training examples. In practice, to compute the gradient descent $\nabla C$ we need to compute the gradients $\nabla C_x$ separately for each training input, x, then average them. This could be a problem when the training inputs is very large, and can take a long time.

### Stochastic Gradient Descent

This idea of stochastic gradient descent can be used to speed up learning. The idea is to estimate hte graidnet $\nabla C$ by computing $\nabla C_x$ for a small sample of randomly selected traning inputs. By averaging over this small sample, it turns our that we can quickly get a good estimate of the true gradient $\nabla C$, and this helps speed up gradient descent, and thus learning. 

In more formal terms, 

we select a small number $m$ of randomly chosen training inputs, labeled $X_1, X_2, ..., X_m$ and refer to them as mini-batch. Provided the sampel size $m$ is large enough, we can expect that the average value of the $\nabla C_{X_j}$ will be roughly equal to the average over all $\nabla C_x$, that is 
$$
\frac{\sum_{j = 1}^m \nabla C_{X_j}}{m} \approx \frac{\sum_x \nabla C_x}{n} = \nabla C
$$
where the second sum is over the entire set of training data. 
$$
\nabla C \approx \frac{1}{m} \sum_{j= 1}^m \nabla C_{X_j}
$$
confirming that we can estimate the overall graident by computing gradients just for the ranodmly chosen mini-batch.

Connecting this notion to neural networks, and suppose that $w_k$ and $b_l$ denote the weights and biases in our neural network. The stochastic gradient descent works by picking out a randomly chosen mini-batch of training inputs, and training with those,

$$
w_k \rightarrow w'_k = w_k - \frac{\eta}{m} \sum_{j} \frac{\delta C_{X_j}}{\delta w_k}
$$
$$
b_l \rightarrow b'_l = b_l - \frac{\eta}{m} \sum_{j} \frac{\delta C_{X_j}}{\delta b_l}
$$
where the sums are over all the training examples $X_j$ in the current mini-batch. Then we pick out another randomly chosen mini-batch and train with those. And so on, until we've exhausted the training inputs, which is said to complete an epoch of training. At that point we start over wit a new training epoch.
